### Vectors and Matrices

```octave
#Matrix A
A = [2,4;5,6];
disp(A)
```

```octave
#Vector v
v = [1;4;5;6;7];
disp(v)
```

```octave
%Matrix Multiplication
B = [1,3,4;2,4,6];
disp(A*B)
```

```octave
%Inverse of A(matrix A must be a non-singular square matrix)
disp(pinv(A))
```

```octave
%Transpose of B
disp(B')
```

```octave
#Matrix C
C = [1,7;2,6];
%Elementwise product(dot product) of A and C
disp('Matrix A')
disp(A)
disp('Matrix C')
disp(C)
disp('Element-wise product')
disp(A.*C)
```

```octave
%to get dot product
disp(sum(sum(A.*C)))
```

```octave
%Element-wise reciprocal of A
disp(1./A)
```

```octave
%Element-wise power 2 of A
disp(A.^2)
```

```octave
%Element-wise functional operation
disp('Log')
disp(log(A))
disp('Exp')
disp(exp(A))
```

```octave
%Magic matrix(matrix with sums along all rows, columns and diagonal are same)
magic_M = magic(5);
disp(magic_M)
```

### Plotting

```octave
t = [0:0.01:0.98];
y = sin(2*pi*8*t);
plot(t,y);
```

```octave
%to plot another plot in the same figure
y1 = cos(2*pi*8*t);
plot(t,y);
hold on;      # <-----
plot(t,y1,'r');
```

```octave
%Label the plot properly
y1 = cos(2*pi*8*t);
plot(t,y);
hold on;
plot(t,y1,'r');
xlabel('time');
ylabel('value');
legend('Sin', 'Cosine');
title('Phase difference of pi/2');
```

```octave
%Subplot(to plot mutliple plots in a grid)
subplot(1,2,1);     #  <---- 1,2 : row x column of grid; and 1 to access first element of grid
plot(t,y1,'r');
subplot(1,2,2);
plot(t,y)
```

```octave
%axis of plot
subplot(1,2,1);     
plot(t,y1,'r');
axis([-0.25 1.25 -1 1]);  #  <---- -025 1.25 : horizontal axis; -1 1 : vertical axis
subplot(1,2,2);
plot(t,y)
axis([-0.25 1.25 -1 1]);  #  <---- -025 1.25 : horizontal axis; -1 1 : vertical axis
```

### Loops

```octave
for i=1:10,
    V(i) = 2.^i;   %indentation is not necessary
end;
disp(V)
```

```octave
i=1
while true,
    if i < 6,
        V(i) = 999;
        i = i+1;
    elseif i < 9,
        V(i) = -999;
        i = i+1;
    else,
        V(i) = 0;
        break;
    end;
end;
disp(V)
```
